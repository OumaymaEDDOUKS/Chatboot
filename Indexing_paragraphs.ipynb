```python
import os
import json
import pprint

from tqdm.notebook import tqdm

from pyserini.index.lucene import LuceneIndexer, IndexReader
from pyserini.pyclass import autoclass

from datasets import load_dataset, Dataset
from datasets.utils.py_utils import convert_file_size_to_int
```

```python

import pandas as pd
import json

# Load your dataset from the JSON file
with open("/content/drive/MyDrive/OOOD.json", 'r') as f:
    dataset = json.load(f)

# Create a DataFrame from the JSON data
data = [{'text': doc['content'], 'label': idx} for idx, doc in enumerate(dataset)]
df = pd.DataFrame(data)


```

```python
import pandas as pd
from datasets import Dataset

# Assuming you have already created the DataFrame df
# df = pd.DataFrame(...)

# Create a Dataset from the DataFrame
dset = Dataset.from_pandas(df)

# (Optional) If you want to rename columns in the Dataset
# If you want to rename the 'text' column to 'contents':
dset = dset.rename_column("text", "contents")

# (Optional) If you want to add a new 'id' column to the Dataset
# Assuming you want to add a 'id' column containing string representations of row indices:
dset = dset.add_column("id", [str(i) for i in range(len(dset))])

# (Optional) If you want to select specific columns in the Dataset
# Assuming you want to keep only the 'id' and 'contents' columns:
dset = dset.select_columns(["id", "contents"])

# Now, your dset is ready, and you can use it for further processing
print(dset)
```

```python
pprint.pprint(dset[0])
```

```python
shard_dir = f"../shards/imdb"
max_shard_size = convert_file_size_to_int("10MB")
dataset_nbytes = dset.data.nbytes
num_shards = int(dataset_nbytes / max_shard_size) + 1
num_shards = max(num_shards, 1)
print(f"Sharding into {num_shards} JSONL files.")
os.makedirs(shard_dir, exist_ok=True)
for shard_index in tqdm(range(num_shards)):
    shard = dset.shard(num_shards=num_shards, index=shard_index, contiguous=True)
    shard.to_json(
        f"{shard_dir}/docs-{shard_index:03d}.jsonl", orient="records", lines=True
    )
```

```python
JIndexCollection = autoclass("io.anserini.index.IndexCollection")
```

```python

indexing_args = [
    "-input",
    shard_dir,
    "-index",
    "../indexes/imdb",
    "-collection",
    "JsonCollection",
    "-threads",
    "28",
    "-language",
    "en",
    "-storePositions",
    "-storeDocvectors",
    "-storeContents",
]
```

```python
%%time
JIndexCollection.main(indexing_args)
```

```python
%%time
batched_streaming_indexer = LuceneIndexer(
    "../indexes/imdb-streaming-batched", threads=28
)
batch_size = 1000
batch = []

for i, row in tqdm(enumerate(dset)):
    batch.append(row)
    if len(batch) >= batch_size:
        batched_streaming_indexer.add_batch_dict(batch)
        batch = []

if len(batch) >= batch_size:
    batched_streaming_indexer.add_batch_dict(batch)

batched_streaming_indexer.close()
```

```python
import os
import shutil

# Specify your new directory path
new_directory = "/content/drive/MyDrive/finish_index"

# Path to the original indexed directory
original_directory = "../indexes/imdb"

# Create the new directory if it doesn't exist
os.makedirs(new_directory, exist_ok=True)

# Copy contents from the original directory to the new directory
for item in os.listdir(original_directory):
    source_item = os.path.join(original_directory, item)
    destination_item = os.path.join(new_directory, item)
    if os.path.isdir(source_item):
        shutil.copytree(source_item, destination_item)
    else:
        shutil.copy2(source_item, destination_item)

print("Contents copied to your new directory:", new_directory)
```
